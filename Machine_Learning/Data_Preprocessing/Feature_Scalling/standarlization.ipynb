```python
df = pd.DataFrame(df)

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the DataFrame
scaler.fit(df)

# Transform the DataFrame
df_standardized = scaler.transform(df)

# Convert the transformed data back to a DataFrame
df_standardized = pd.DataFrame(df_standardized, columns=df.columns)

print(df_standardized)
```

```python
# Sample DataFrame
data = {'A': [10, 20, 30],
        'B': [40, 50, 60],
        'C': ['a', 'b', 'c']}
df = pd.DataFrame(data)

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the specific columns you want to standardize
columns_to_standardize = ['A', 'B']
df_standardized = df.copy()
df_standardized[columns_to_standardize] = scaler.fit_transform(df[columns_to_standardize])

print(df_standardized)
```
