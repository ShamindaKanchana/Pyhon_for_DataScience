```python
import pandas as pd
import seaborn as sns  
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv", encoding='latin1')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
#To find outliers
sns.boxplot(x="loan_amnt",data=dataset)
```

```python
#min=Q1-(1.5XIQR)
#max=Q3+(1.5XIQR)
dataset.shape
```

```python
q1=dataset["loan_amnt"].quantile(0.25)
q3=dataset["loan_amnt"].quantile(0.75)
IQR=q3-q1

```

```python
min_range=q1-(1.5*IQR)
max_range=q3+(1.5*IQR)
min_range,max_range
```

```python
new_dataset=dataset[dataset["loan_amnt"]<=max_range]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="loan_amnt",data=new_dataset)
```

```python
#Using zscore method remove outliers
dataset.isnull().sum()
dataset.describe()
```

```python
sns.boxplot(x="revol_bal",data=new_dataset,color="red")
```

```python
#through zscore way 
min_range=dataset["loan_amnt"].mean()-(3*dataset["loan_amnt"].std())
max_range=dataset["loan_amnt"].mean()+(3*dataset["loan_amnt"].std())
```

```python
min_range,max_range
```

```python
new_data=dataset[dataset["loan_amnt"]<=max_range]
sns.boxplot(x="loan_amnt",data=new_data)
```

```python
z_score=(dataset["loan_amnt"]-dataset["loan_amnt"].mean())/(dataset["loan_amnt"].std())
```

```python
z_score>3
```

```python
dataset["z_score"]=z_score
dataset
```

```python
#remmove outliers
dataset[dataset["z_score"]<3]
```
