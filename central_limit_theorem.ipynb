```python
import numpy as np
import seaborn  as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
pop_data=[np.random.randint(10,100) for i in range(10000)]
pop_table=pd.DataFrame({"Pop_data":pop_data})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Pop_data",data=pop_table)
plt.show()
```

```python
#take sample from the above population data 
sample_mean=[]
for no_sample in range(50):
    sample_data=[]
    for i in range(500):
        #np.random.choice(pop_data) take random data from the population
        sample_data.append(np.random.choice(pop_data)) #insert each of sample values to sample_data array
    sample_mean.append(np.mean(sample_data))    
    
```

```python
sample_M=pd.DataFrame({"Sample_Mean":sample_mean})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Sample_Mean",data=sample_M)
plt.show()
```
